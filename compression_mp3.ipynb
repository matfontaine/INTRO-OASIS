```python
import numpy as np
import scipy.fft
import scipy.io
import matplotlib.pyplot as plt
import IPython.display as ipd
import soundfile as sf
```

```python
fichier = 'gamme_guitare_reel_do_majeur.wav'
x, Fe = sf.read(fichier)
x = x[:,0]
```

```python
plt.plot(np.arange(len(x))/Fe, x)
plt.xlabel('t en s')
plt.ylabel('x(t)')
```

```python
spectre = np.fft.fft(x)
plt.semilogy(np.arange(len(spectre))*Fe/len(spectre), np.abs(spectre))
plt.xlabel('f en Hz')
plt.ylabel('$\hat x(f)$')
[ybas, yhaut] = plt.ylim()
```

```python
dct = scipy.fft.dct(x)
plt.semilogy(np.arange(len(dct))*Fe/len(dct), np.abs(dct))
plt.xlabel('f en Hz')
plt.ylabel('$\hat x(f)$')
[ybas, yhaut] = plt.ylim()
```

```python
compression_level = 0.05
nb_bands = 1

fft = scipy.fft.fft(x)
comp_fft = 0 * fft
for band in range(nb_bands):
    freqs = np.arange((len(fft)*band) // nb_bands,(len(fft)*(band+1)) // nb_bands)
    fft_bin = fft[freqs]
    sorted_bin = np.argsort(np.abs(fft_bin))
    compressed_bin = 0 * fft_bin
    compressed_bin[sorted_bin[-int(compression_level*len(fft_bin)):]] = fft_bin[sorted_bin[-int(compression_level*len(fft_bin)):]]
    #if np.linalg.norm(compressed_bin) > 0:
    #    compressed_bin = np.linalg.norm(dct_bin,2)/np.linalg.norm(compressed_bin) * compressed_bin
    comp_fft[freqs] = compressed_bin
    
```

```python
compression_level = 0.05
nb_bands = 10

dct = scipy.fft.dct(x)
comp_dct = 0 * dct
for band in range(nb_bands):
    freqs = np.arange((len(dct)* (2**band-1)) // (2**nb_bands-1),(len(dct)*(2**(band+1)-1)) // (2**nb_bands-1))
    dct_bin = dct[freqs]
    sorted_bin = np.argsort(np.abs(dct_bin))
    compressed_bin = 0 * dct_bin
    compressed_bin[sorted_bin[-int(compression_level*len(dct_bin)):]] = dct_bin[sorted_bin[-int(compression_level*len(dct_bin)):]]
    #if np.linalg.norm(compressed_bin) > 0:
    #    compressed_bin = np.linalg.norm(dct_bin,2)/np.linalg.norm(compressed_bin) * compressed_bin
    comp_dct[freqs] = compressed_bin
    print(freqs[0], freqs[-1])
```

```python
comp_x = scipy.fft.idct(comp_dct)
comp_x_fft = scipy.fft.ifft(comp_fft)
```
